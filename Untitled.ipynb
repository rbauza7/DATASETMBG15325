```r
# Cargar librerías necesarias
library(readr)
library(tidyr)
library(dplyr)

# URL RAW del archivo en GitHub
url <- "https://raw.githubusercontent.com/rbauza7/malaria_frecuencia_semanal/refs/heads/main/malaria_frecuencia_semanal2.csv"

# Leer el CSV
malaria_data <- read_csv(url)

# Ver los primeros registros
head(malaria_data)
```

Para esto vamos a modelar la evolución global (todas las semanas de todos los años) para identificar si existe tendencia descendente general, usando regresión lineal simple sobre el tiempo.

 Pasos del análisis:

Creamos una variable Semana_Absoluta que numere cada semana desde la primera del 2020 hasta la última del 2024.

Ajustamos un modelo de regresión lineal:

ini
Copiar
Editar
Casos = β0 + β1*(Semana_Absoluta)
Así evaluamos si la pendiente es negativa (tendencia descendente).

Graficamos la tendencia.

```r
# Ordenar los datos por año y semana
malaria_long <- malaria_long %>%
  arrange(Anio, Semana)

# Crear una variable de semana absoluta
malaria_long <- malaria_long %>%
  mutate(Semana_Absoluta = row_number())

# Ajustar modelo de regresión lineal
modelo_tendencia <- lm(Casos ~ Semana_Absoluta, data = malaria_long)

# Mostrar resumen del modelo
summary(modelo_tendencia)
```

ué mirarás en la salida?

Estimate de Semana_Absoluta: Si es negativo y significativo (p < 0.05), confirma tendencia descendente.

R-squared: Proporción de varianza explicada por el modelo.

 Modelo ajustado:

\text{Casos} = 47.16 - 0.169 \times \text{Semana_Absoluta}
🔍 Interpretación de los coeficientes:

Intercepto (47.16):
El número estimado de casos en la semana 1 de la serie (inicio de 2020).

Pendiente (-0.169):
Cada semana que pasa, el número promedio de casos disminuye en aproximadamente 0.169 casos.
Esto confirma una tendencia descendente sostenida en el tiempo.

✅ Significancia estadística:

El p-valor de la pendiente es <2e-16, extremadamente significativo.

F-statistic = 124.5, p<2e-16: el modelo es globalmente significativo.

📊 Bondad de ajuste:

R-squared = 0.325
(La tendencia explica el 32.5% de la variabilidad de los casos semanales.)
Esto es moderado: hay otros factores (estacionalidad, brotes) que también influyen.

🔹 En resumen:
Tu análisis confirma que entre 2020 y 2024 hubo una reducción sistemática y significativa de los casos de malaria.

Mensaje importante:

"I can't handle data with frequency greater than 24. Seasonality will be ignored."

Esto ocurre porque ets() no soporta frecuencias mayores a 24 (tu serie tiene frecuencia 52 semanas/año).

Resultado:
El modelo que se ha ajustado es ETS(A,N,N):

A: Error aditivo

N: Sin tendencia

N: Sin estacionalidad

En otras palabras, hizo un suavizamiento exponencial sin tendencia ni estacionalidad, solo nivel.

 Ajustar un ARIMA automáticamente con auto.arima().

Ahora vas a ajustar un ARIMA de manera automática usando auto.arima(), que selecciona el mejor modelo según AICc, BIC y el comportamiento de la serie.

1. Asegúrate de tener forecast cargado

🔹 2. Ajustar el modelo ARIMA automáticamente

```r
# Ajustar ARIMA automático
modelo_arima <- auto.arima(ts_malaria)

# Mostrar resumen del modelo
summary(modelo_arima)
```

Modelo seleccionado automáticamente:
ARIMA(5,1,4)(1,0,0)[52] con drift

Esto significa:

p=5 (5 términos autoregresivos)

d=1 (una diferencia para hacer la serie estacionaria)

q=4 (4 términos de media móvil)

P=1 (un componente autoregresivo estacional)

D=0 (no diferencia estacional)

Q=0 (sin media móvil estacional)

S=52 (periodo estacional = 52 semanas)

drift (término de tendencia constante)

🔍 Interpretación de los componentes:

El modelo reconoce que hay persistencia fuerte (p=5) y fluctuaciones de corto plazo (q=4).

La diferencia de orden 1 elimina la tendencia para que la serie sea estacionaria.

El componente estacional AR(1) captura la correlación a 52 semanas (un ciclo anual moderado).

El drift permite capturar una pendiente media residual.

✅ Medidas de ajuste:

AIC = 2199.83

AICc = 2201.1

BIC = 2242.5
Son valores esperados para una serie larga con alta variabilidad.

RMSE ≈ 16.0
El error cuadrático medio de predicción en el entrenamiento.

✅ ACF1 residual ≈ 0.018
Muy bajo, indica que los residuos están casi no correlacionados (buena señal).

🔹 Siguiente paso clave:
Vamos a ver si los residuos cumplen con “ruido blanco”.



```r
# Diagnóstico de residuos
checkresiduals(modelo_arima)
```

🔍 Gráfico inferior izquierdo (ACF de residuos):

La mayoría de barras están dentro de las bandas azules, indicando que los residuos son casi ruido blanco.

Este modelo ARIMA(5,1,4)(1,0,0)[52] con drift se ajusta bastante bien.

```r
# Librería de gráficos
library(ggplot2)

ggplot(pred_comparacion, aes(x = Semana)) +
  # Bandas STLF
  geom_ribbon(aes(ymin = Lower80_STLF, ymax = Upper80_STLF),
              fill = "lightblue", alpha = 0.3) +
  # Línea STLF
  geom_line(aes(y = Pred_STLF, color = "STLF"), size = 1) +
  # Bandas ARIMA
  geom_ribbon(aes(ymin = Lower80_ARIMA, ymax = Upper80_ARIMA),
              fill = "lightgreen", alpha = 0.3) +
  # Línea ARIMA
  geom_line(aes(y = Pred_ARIMA, color = "ARIMA"), size = 1) +
  labs(
    title = "Comparación de proyecciones de casos semanales de malaria",
    subtitle = "STLF (Azul) vs ARIMA (Verde)",
    x = "Semana proyectada",
    y = "Número de casos",
    color = "Modelo"
  ) +
  scale_color_manual(values = c("STLF" = "blue", "ARIMA" = "darkgreen")) +
  theme_minimal(base_size = 14)
```

Interpretación del gráfico:

Línea azul (STLF):

Más fluctuante y con picos, porque STLF retiene parte de la estacionalidad y de la variabilidad del ciclo semanal.

Intervalo de confianza (banda azul) algo más ancho al inicio, indicando mayor incertidumbre en la tendencia reciente.

Línea verde (ARIMA):

Más suave y estable.

Muestra una proyección baja y plana.

Banda de confianza (verde) relativamente constante.



 ¿Por qué se diferencian?

STLF hace una descomposición (tendencia + estacionalidad) y luego proyecta cada parte.

Captura fluctuaciones más marcadas.

ARIMA modela todo en conjunto con diferenciado y suavizado.

Tiende a proyecciones más conservadoras.

Qué puedes decir en un informe o publicación:

“El modelo STLF refleja mayor sensibilidad a la variabilidad histórica y retiene fluctuaciones estacionales, mientras que ARIMA genera un pronóstico más suavizado y conservador. La elección del método dependerá del nivel de detalle deseado y del riesgo tolerado en la planificación.”
