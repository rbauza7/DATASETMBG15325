```r
# Cargar librer√≠as necesarias
library(readr)
library(tidyr)
library(dplyr)

# URL RAW del archivo en GitHub
url <- "https://raw.githubusercontent.com/rbauza7/malaria_frecuencia_semanal/refs/heads/main/malaria_frecuencia_semanal2.csv"

# Leer el CSV
malaria_data <- read_csv(url)

# Ver los primeros registros
head(malaria_data)
```

Para esto vamos a modelar la evoluci√≥n global (todas las semanas de todos los a√±os) para identificar si existe tendencia descendente general, usando regresi√≥n lineal simple sobre el tiempo.

 Pasos del an√°lisis:

Creamos una variable Semana_Absoluta que numere cada semana desde la primera del 2020 hasta la √∫ltima del 2024.

Ajustamos un modelo de regresi√≥n lineal:

ini

Casos = Œ≤0 + Œ≤1*(Semana_Absoluta)
As√≠ evaluamos si la pendiente es negativa (tendencia descendente).

Graficamos la tendencia.

```r
# Ordenar los datos por a√±o y semana
malaria_long <- malaria_long %>%
  arrange(Anio, Semana)

# Crear una variable de semana absoluta
malaria_long <- malaria_long %>%
  mutate(Semana_Absoluta = row_number())

# Ajustar modelo de regresi√≥n lineal
modelo_tendencia <- lm(Casos ~ Semana_Absoluta, data = malaria_long)

# Mostrar resumen del modelo
summary(modelo_tendencia)
```

u√© mirar√°s en la salida?

Estimate de Semana_Absoluta: Si es negativo y significativo (p < 0.05), confirma tendencia descendente.

R-squared: Proporci√≥n de varianza explicada por el modelo.

 Modelo ajustado:

\text{Casos} = 47.16 - 0.169 \times \text{Semana_Absoluta}
üîç Interpretaci√≥n de los coeficientes:

Intercepto (47.16):
El n√∫mero estimado de casos en la semana 1 de la serie (inicio de 2020).

Pendiente (-0.169):
Cada semana que pasa, el n√∫mero promedio de casos disminuye en aproximadamente 0.169 casos.
Esto confirma una tendencia descendente sostenida en el tiempo.

‚úÖ Significancia estad√≠stica:

El p-valor de la pendiente es <2e-16, extremadamente significativo.

F-statistic = 124.5, p<2e-16: el modelo es globalmente significativo.

üìä Bondad de ajuste:

R-squared = 0.325
(La tendencia explica el 32.5% de la variabilidad de los casos semanales.)
Esto es moderado: hay otros factores (estacionalidad, brotes) que tambi√©n influyen.

üîπ En resumen:
Tu an√°lisis confirma que entre 2020 y 2024 hubo una reducci√≥n sistem√°tica y significativa de los casos de malaria.

```r
# Crear serie temporal semanal con frecuencia 52
ts_malaria <- ts(
  malaria_long$Casos,
  frequency = 52,
  start = c(2020, 1)
)

# Verificar resumen
summary(ts_malaria)
```

```r
# Ajustar modelo ETS
modelo_ets <- ets(ts_malaria)

# Ver resumen del modelo
summary(modelo_ets)
```

Mensaje importante:

"I can't handle data with frequency greater than 24. Seasonality will be ignored."

Esto ocurre porque ets() no soporta frecuencias mayores a 24 (tu serie tiene frecuencia 52 semanas/a√±o).

Resultado:
El modelo que se ha ajustado es ETS(A,N,N):

A: Error aditivo

N: Sin tendencia

N: Sin estacionalidad

En otras palabras, hizo un suavizamiento exponencial sin tendencia ni estacionalidad, solo nivel.

Interpretaci√≥n r√°pida cuando lo veas:

ACF te muestra qu√© tan correlacionados est√°n los datos con sus rezagos.

Si hay barras altas y significativas (fuera de las bandas azules), significa dependencia temporal.

Picos a lag=52 indican estacionalidad anual.

PACF te dice cu√°ntos rezagos explican esa correlaci√≥n de manera directa.

Es clave para elegir el par√°metro p en ARIMA.



‚úÖ ACF (Autocorrelaci√≥n):

Lag 1: alt√≠simo (~0.5), muestra fuerte dependencia de corto plazo.

La autocorrelaci√≥n decae gradualmente, pero sigue por encima de las bandas azules hasta aproximadamente lag 20‚Äì25, indicando persistencia temporal.

Este patr√≥n de decaimiento lento sugiere una combinaci√≥n de tendencia + estacionalidad difusa.

No se ve un pico clar√≠simo en lag=52, as√≠ que la estacionalidad anual no es dominante o est√° enmascarada por la variabilidad.

‚úÖ PACF (Autocorrelaci√≥n Parcial):

Lag 1: significativamente alto (~0.4), confirmando que el rezago inmediato explica gran parte de la dependencia.

A partir de lag 2, las barras caen r√°pido y oscilan cerca de 0, aunque algunos rezagos individuales quedan por encima de la banda.

Este patr√≥n sugiere que un AR(1) podr√≠a capturar buena parte de la correlaci√≥n inmediata.

üîç Interpretaci√≥n global:

Tu serie es altamente autocorrelacionada en el corto plazo (efecto inercia semana a semana).

La estacionalidad es menos clara (probablemente por fluctuaciones impredecibles).

Hay evidencia de que un modelo ARIMA con bajo orden AR y posible componente MA podr√≠a ser adecuado.

Ahora vas a ajustar un ARIMA de manera autom√°tica usando auto.arima(), que selecciona el mejor modelo seg√∫n AICc, BIC y el comportamiento de la serie.

üîπ 2. Ajustar el modelo ARIMA autom√°ticamente

```r
# Ajustar ARIMA autom√°tico
modelo_arima <- auto.arima(ts_malaria)

# Mostrar resumen del modelo
summary(modelo_arima)
```

Modelo seleccionado autom√°ticamente:
ARIMA(5,1,4)(1,0,0)[52] con drift

Esto significa:

p=5 (5 t√©rminos autoregresivos)

d=1 (una diferencia para hacer la serie estacionaria)

q=4 (4 t√©rminos de media m√≥vil)

P=1 (un componente autoregresivo estacional)

D=0 (no diferencia estacional)

Q=0 (sin media m√≥vil estacional)

S=52 (periodo estacional = 52 semanas)

drift (t√©rmino de tendencia constante)

üîç Interpretaci√≥n de los componentes:

El modelo reconoce que hay persistencia fuerte (p=5) y fluctuaciones de corto plazo (q=4).

La diferencia de orden 1 elimina la tendencia para que la serie sea estacionaria.

El componente estacional AR(1) captura la correlaci√≥n a 52 semanas (un ciclo anual moderado).

El drift permite capturar una pendiente media residual.

‚úÖ Medidas de ajuste:

AIC = 2199.83

AICc = 2201.1

BIC = 2242.5
Son valores esperados para una serie larga con alta variabilidad.

RMSE ‚âà 16.0
El error cuadr√°tico medio de predicci√≥n en el entrenamiento.

‚úÖ ACF1 residual ‚âà 0.018
Muy bajo, indica que los residuos est√°n casi no correlacionados (buena se√±al).

1. Crear un data.frame combinado con ambas predicciones

```r
# DataFrame STLF
pred_stlf_df <- data.frame(
  Semana = 1:length(proyeccion_stlf$mean),
  Pred_STLF = as.numeric(proyeccion_stlf$mean),
  Lower80_STLF = as.numeric(proyeccion_stlf$lower[,1]),
  Upper80_STLF = as.numeric(proyeccion_stlf$upper[,1])
)

# DataFrame ARIMA
pred_arima_df <- data.frame(
  Semana = 1:length(proyeccion_arima$mean),
  Pred_ARIMA = as.numeric(proyeccion_arima$mean),
  Lower80_ARIMA = as.numeric(proyeccion_arima$lower[,1]),
  Upper80_ARIMA = as.numeric(proyeccion_arima$upper[,1])
)

# Combinar en un solo dataframe
pred_comparacion <- merge(pred_stlf_df, pred_arima_df, by = "Semana")
```

2. Graficar ambas series con colores y leyenda clara

```r
# Librer√≠a de gr√°ficos
library(ggplot2)

ggplot(pred_comparacion, aes(x = Semana)) +
  # Bandas STLF
  geom_ribbon(aes(ymin = Lower80_STLF, ymax = Upper80_STLF),
              fill = "lightblue", alpha = 0.3) +
  # L√≠nea STLF
  geom_line(aes(y = Pred_STLF, color = "STLF"), size = 1) +
  # Bandas ARIMA
  geom_ribbon(aes(ymin = Lower80_ARIMA, ymax = Upper80_ARIMA),
              fill = "lightgreen", alpha = 0.3) +
  # L√≠nea ARIMA
  geom_line(aes(y = Pred_ARIMA, color = "ARIMA"), size = 1) +
  labs(
    title = "Comparaci√≥n de proyecciones de casos semanales de malaria",
    subtitle = "STLF (Azul) vs ARIMA (Verde)",
    x = "Semana proyectada",
    y = "N√∫mero de casos",
    color = "Modelo"
  ) +
  scale_color_manual(values = c("STLF" = "blue", "ARIMA" = "darkgreen")) +
  theme_minimal(base_size = 14)
```

Interpretaci√≥n del gr√°fico:

L√≠nea azul (STLF):

M√°s fluctuante y con picos, porque STLF retiene parte de la estacionalidad y de la variabilidad del ciclo semanal.

Intervalo de confianza (banda azul) algo m√°s ancho al inicio, indicando mayor incertidumbre en la tendencia reciente.

L√≠nea verde (ARIMA):

M√°s suave y estable.

Muestra una proyecci√≥n baja y plana.

Banda de confianza (verde) relativamente constante.



 ¬øPor qu√© se diferencian?

STLF hace una descomposici√≥n (tendencia + estacionalidad) y luego proyecta cada parte.

Captura fluctuaciones m√°s marcadas.

ARIMA modela todo en conjunto con diferenciado y suavizado.

Tiende a proyecciones m√°s conservadoras.
